```python
# Filter low-occurrence labels
threshold = int(len(df) * 0.00467)
rare_topics = [key for key, value in sorted_topic.items() if value < threshold]
```

```python
len(rare_topics)
```

```python
print(rare_topics)
```

```python
topics_list = df.topics.to_list()

revised_topic_list = []
indices_to_drop = []

for idx, topics in enumerate(topics_list):
  revised_topics = [topic.strip() for topic in topics.split(',') if topic.strip() not in rare_topics]
  if len(revised_topics) == 0:
    indices_to_drop.append(idx)
  else:
    revised_topic_list.append(revised_topics)
```

```python
len(revised_topic_list)
```

```python
df = df.drop(indices_to_drop).reset_index(drop=True)
# Save cleaned topics back to dataframe
df["topics"] = revised_topic_list
print(df.shape)
```

```python
df.head()
```

```python
revised_topics_list = df.topics.to_list()
revised_topic_count = {}
for topics in revised_topics_list:
  topic_list = topics
  for topic in topic_list:
    if topic in revised_topic_count.keys():
      revised_topic_count[topic] += 1
    else:
      revised_topic_count[topic] = 1
print(f"Number of topics: {len(revised_topic_count)}")
print(dict(sorted(revised_topic_count.items(), key=lambda x:x[1], reverse=True)))
```

```python
print(revised_topic_count.items())
```

```python
encode_topics_types = {key: idx for idx, (key, value) in enumerate(revised_topic_count.items())}
with open("topic_encoded.json", "w") as fp:
  json.dump(encode_topics_types, fp)
```

```python
len(encode_topics_types)
```

```python
# We need this because for multilabel classification all topics have possibility to be present in the predictions
categorical_topic_list = []
revised_topics_list = df.topics.to_list()

for revised_topics in revised_topics_list:
  categorical_list = [0] * len(encode_topics_types)
  for topic in revised_topics:
    topic_type_index = encode_topics_types[topic]
    categorical_list[topic_type_index] = 1
  categorical_topic_list.append(categorical_list)
```

```python
#print(categorical_topic_list[20000])
```

```python
df['topic_category'] = categorical_topic_list
```
